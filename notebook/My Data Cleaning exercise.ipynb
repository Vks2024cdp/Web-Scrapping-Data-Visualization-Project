```python
import pandas as pd
import re
```

```python
##Data Cleaning Process:
##Harvey Norman dataset  

HN_df = pd.read_csv('C:/Users/User/5m-data-2.4-data-extraction-scraping/notebooks/harvey_norman_data/harvey_norman_products_20250316-114221.csv')

# Display basic information about the dataset
print("Dataset Overview:")
print(f"Number of products: {len(HN_df)}")
print(HN_df.head())
print("\nData Information:")
print(HN_df.info())
print("\nSummary Statistics:")
print(HN_df.describe())

# Identify rows where the value isn't numeric for original_price
invalid_values = HN_df.loc[
    ~HN_df['original_price'].astype(str).str.replace(r'[S$,]', '', regex=True).str.replace('.', '', regex=True).str.isnumeric(),
    'original_price'
]

HN_df['original_price'] = HN_df['original_price'].astype(str).str.replace(r'[S$,]', '', regex=True)
HN_df['original_price'] = pd.to_numeric(HN_df['original_price'], errors='coerce')
HN_df['original_price'].fillna(0, inplace=True)


# Identify rows where the value isn't numeric for current price
invalid_values = HN_df.loc[
    ~HN_df['current_price'].astype(str).str.replace(r'[S$,]', '', regex=True).str.replace('.', '', regex=True).str.isnumeric(),
    'current_price'
]

HN_df['current_price'] = HN_df['current_price'].astype(str).str.replace(r'[S$,]', '', regex=True)
HN_df['current_price'] = pd.to_numeric(HN_df['current_price'], errors='coerce')
HN_df['current_price'].fillna(0, inplace=True)


# If discount_rate is given as percentage string (e.g., '20%'), convert to float
if HN_df['discount_percentage'].dtype == 'object':
    HN_df['discount_percentage'] = HN_df['discount_percentage'].str.replace('%', '').astype(float)

HN_df['discount_percentage'].fillna(0, inplace=True)

# Calculate absolute discount amount
HN_df['discount_amount'] = HN_df['original_price'] - HN_df['current_price']

## Replace negative values with zero
HN_df.loc[HN_df['discount_amount'] < 0, 'discount_amount'] = 0

# Print updated DataFrame
print(HN_df.head())
```

```python
### Find Category for Harvey Norman Products

# Define more comprehensive category keywords for better detection
common_categories = {
    'TV': ['tv', 'television', 'smart tv', 'oled', 'qled', 'led tv', 'uhd', '4k', '8k', 'hdtv', 'curved', 'android tv', 'webos'],
    'Monitor': ['monitor', 'gaming monitor', 'ultrawide', 'display', 'curved monitor', 'lcd', 'led monitor'],
    'Refrigerator': ['fridge', 'refrigerator', 'refrigerators', 'freezer', 'cooler', 'mini fridge', 'french door', 'side by side'],
    'Washing Machine': ['washer', 'washing machine', 'dryer', 'laundry', 'front load', 'top load', 'combo'],
    'Dishwasher': ['dishwasher', 'dish washer'],
    'Laptop': ['laptop', 'notebook', 'chromebook', 'macbook', 'ultrabook', 'gaming laptop' , 'imac'],
    'Desktop': ['desktop', 'pc', 'computer', 'all-in-one', 'tower', 'workstation', 'gaming pc'],
    'Smartphone': ['phone', 'smartphone', 'iphone', 'galaxy', 'mobile', 'cell phone'],
    'Tablet': ['tablet', 'ipad', 'galaxy tab', 'slate', 'e-reader', 'kindle'],
    'Audio': ['speaker', 'headphone', 'earphone', 'earbuds', 'soundbar', 'audio', 'bluetooth', 'wireless speaker', 'subwoofer', 'woofer', 'sound bar'],
    'Kitchen Appliance': ['microwave', 'oven', 'cooktop', 'stove', 'blender', 'juicer', 'mixer', 'coffee', 'toaster', 'kettle', 'hood', 'maker'
                          'espresso', 'food processor', 'air fryer', 'slow cooker', 'rice cooker', 'pressure cooker', 'frypan', 'water purifier', 'saucepan', 
                         'grinder', 'chopper', 'burner', 'hob', 'slicer', 'induction cooker'],
    'Small Appliance': ['iron', 'vacuum', 'steamer', 'fan', 'heater', 'humidifier', 'dehumidifier', 'air purifier', 'water dispenser'],
    'Vacuum': ['vacuum', 'cleaner', 'steam mop', 'floor cleaner', 'robot vacuum', 'stick vacuum', 'handheld vacuum'],
    'Gaming': ['console', 'playstation', 'xbox', 'nintendo', 'gaming', 'controller', 'ps5', 'ps4', 'vr', 'oculus'],
    'Camera': ['camera', 'dslr', 'mirrorless', 'lens', 'action cam', 'webcam', 'security camera', 'gopro'],
    'Furniture': ['sofa', 'couch', 'bed', 'mattress', 'chair', 'table', 'desk', 'wardrobe', 'dresser', 'bookshelf', 'ottoman'],
    'Air Conditioner': ['air conditioner', 'air con', 'ac', 'split system', 'aircon', 'portable ac', 'heating', 'cooling'],
    'Printer': ['printer', 'scanner', 'inkjet', 'laser', 'all-in-one printer', 'multifunction'],
    'Wearable': ['watch', 'smartwatch', 'fitness tracker', 'band', 'apple watch', 'garmin', 'fitbit'],
    'Storage': ['ssd', 'hard drive', 'external', 'nas', 'thumb drive', 'usb drive', 'memory card', 'microsd'],
    'Networking': ['router', 'modem', 'mesh', 'wifi', 'network', 'switch', 'access point', 'extender'],
    'Home Entertainment': ['projector', 'home theater', 'blu-ray', 'dvd player', 'amplifier', 'receiver', 'media player'],
    'Accessories': ['keyboard', 'mouse', 'webcam', 'headset', 'charger', 'case', 'stand', 'mount', 'cable', 'adapter']
}

def improved_extract_category(product_name):
    """
    Extract product category by searching the entire product name for category keywords.
    More thorough approach that doesn't rely on the second word as fallback.
    """
    product_name = str(product_name).lower()
    
    # Try to find a matching category using our comprehensive keyword list
    for category, keywords in common_categories.items():
        for keyword in keywords:
            # Use word boundary to avoid partial matches (e.g., "phone" in "headphone")
            if re.search(r'\b' + re.escape(keyword) + r'\b', product_name):
                return category
    
    # Remove the brand name from the product name to look at the key descriptors
    words = product_name.split(' ')
    if len(words) > 1:
        remaining_text = ' '.join(words[1:])
        
        # Look for common product type words in what's left
        product_types = ['tv', 'phone', 'laptop', 'monitor', 'fridge', 'oven', 'washer', 'dryer', 
                         'console', 'speaker', 'camera', 'printer', 'tablet', 'watch']
        
        for product_type in product_types:
            if product_type in remaining_text:
                # Map common product types to our categories
                if product_type == 'tv':
                    return 'TV'
                elif product_type == 'phone':
                    return 'Smartphone'
                elif product_type == 'monitor':
                    return 'Monitor'
                elif product_type in ['fridge']:
                    return 'Refrigerator'
                elif product_type in ['washer', 'dryer']:
                    return 'Washing Machine'
                elif product_type in ['oven']:
                    return 'Kitchen Appliance'
                else:
                    # Capitalize the product type for a reasonable category name
                    return product_type.capitalize()
        
        # If we reach here, no specific product type was found
        # Let's check for dimensions which often indicate TVs or monitors
        if re.search(r'\b\d{2}[\"\']|\b\d{2} inch', product_name):
            if 'gaming' in product_name:
                return 'Monitor'
            else:
                return 'TV'
    
    # If nothing else works, return "Other"
    return "Other"

# Apply the improved category extraction
HN_df['category'] = HN_df['product_name'].apply(improved_extract_category)
CS_df['category'] = CS_df['product_name'].apply(improved_extract_category)


# Display the results
print("Extracted Brand and Category with Improved Method:")
print(HN_df[['product_name', 'brand', 'category']].head(10))
print(CS_df[['product_name', 'brand', 'category']].head(10))


# Analyze the distribution of categories
print("\nCategory Distribution:")
HN_category_counts = HN_df['category'].value_counts()
CS_category_counts = CS_df['category'].value_counts()
print(HN_category_counts)
print(CS_category_counts)


# Check how many products are categorized as "Other"
other_count = len(HN_df[HN_df['category'] == 'Other'])
print(f"\nProducts categorized as 'Other': {other_count} ({other_count/len(HN_df)*100:.2f}%)")

# If there are many "Other" categories, look at some examples to refine our categories
if other_count > 0:
    print("\nSample of products categorized as 'Other':")
    other_products = HN_df[HN_df['category'] == 'Other']['product_name'].sample(min(30, other_count)).tolist()
    for product in other_products:
        print(f"- {product}")

```

```python
def extract_product_type(product_name):
    """
    Identify the product type based on keywords from the common_categories dictionary.
    """
    product_name = str(product_name).lower()
    
    for category, keywords in common_categories.items():
        for keyword in keywords:
            if re.search(r'\b' + re.escape(keyword) + r'\b', product_name):
                return keyword  # Return the specific product type (matched keyword)
    
    return "Other"

# Apply the function to create the 'product_type' column
HN_df['product_type'] = HN_df['product_name'].apply(extract_product_type)
CS_df['product_type'] = CS_df['product_name'].apply(extract_product_type)


# Display the results
print(HN_df[['product_name', 'product_type']])
print(CS_df[['product_name', 'product_type']])
```
